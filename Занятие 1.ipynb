## Задача 1: проверка безопасности уровня радиации

Описание: создайте программу, которая определяет, безопасен ли уровень радиации. Пользователь вводит текущий уровень радиации, и программа выводит "Безопасно", "Предупреждение" или "Опасно" в зависимости от заданных порогов.


```python
radiation_level = int(input('Введите уровень радиации'))

print('Уровень радиации:', radiation_level)

if radiation_level < 0:
    raise ValueError('Значение не может быть отрицательным')

if radiation_level <= 40:
    print('Безопасно')
elif radiation_level > 40 and radiation_level < 50:
    print('Предупреждение')
else:
    print('Опасно')
```

## Задача 2: расчет времени до следующего техобслуживания

Описание: напишите программу, которая вычисляет количество дней до следующего технического обслуживания реактора. Пользователь вводит дату последнего обслуживания и периодичность в днях.

```python
from datetime import timedelta, datetime

# текущая дата
today = datetime.now()

print('Сегодня', today.date())

# дата последнего обслуживания
last_date = datetime.strptime(input(
    'Введите дату последнего обслуживания в формате year-month-day'), '%Y-%m-%d')

print('Дата последнего обслуживания', last_date.date())

# период до следующего обслуживания
period = timedelta(int(input('Период обслуживания')))

print(f'Период обслуживания {period.days} дней')

# Сколько дней осталось
days_until_service = (last_date + period) - today

print('--------------------------------------------')

if days_until_service < timedelta(0):
    raise ValueError('Дата обслуживания уже пропущена!')

print('Осталось {} дней'.format(days_until_service.days))
```

## Задача 3*: оптимизация загрузки топлива в реактор

Разработайте программу для определения оптимальной схемы загрузки топлива в ядерный реактор. Имеется несколько типов топливных стержней, каждый с разной степенью эффективности и сроком службы. Программа должна принимать количество доступных стержней каждого типа и требуемое общее количество стержней для загрузки. Задача - распределить стержни так, чтобы максимизировать общую эффективность реактора, учитывая, что срок службы отдельных стержней не должен превышать заданное значение. Реализуйте алгоритм, который вычисляет оптимальное сочетание стержней, исходя из их характеристик и общих ограничений.

**Входные Данные:**

1. Типы Топливных Стержней: Представлены с их эффективностью и сроком службы.

- Тип A: Эффективность 1.5, Срок службы 5 лет

- Тип B: Эффективность 1.2, Срок службы 3 года

- Тип C: Эффективность 1.0, Срок службы 4 года

2. Доступное Количество Стержней Каждого Типа:

- Тип A: 10 стержней

- Тип B: 15 стержней

- Тип C: 20 стержней

3. Требуемое Общее Количество Стержней для Загрузки: 30 стержней

4. Максимальный Срок Службы Реактора: 4 года

**Задача:**

Определить, сколько стержней каждого типа следует использовать, чтобы максимизировать общую эффективность реактора, при этом соблюдая максимальный срок службы реактора.

**Пример Решения:**

1. Расчет Эффективности для Каждого Типа Стержней:

- Тип A: 1.5 * 10 = 15

- Тип B: 1.2 * 15 = 18

- Тип C: 1.0 * 20 = 20

2. Определение Сочетания Стержней:

- Из-за ограничения срока службы реактора, стержни типа B не могут быть использованы (их срок службы 3 года, а максимальный срок службы реактора 4 года).

- Используем максимально возможное количество стержней типа A (10 стержней) для максимизации эффективности.

- Оставшиеся 20 стержней будут типа C.

**Выходные Данные:**

- Использованные Стержни:

  - Тип A: 10 стержней

  - Тип C: 20 стержней

- Общая Эффективность Реактора: Рассчитывается на основе выбранных стержней.

```python
# Данные заданы программно

types = ['A', 'B', 'C']
efficiency = [1.5, 1.2, 1.0]
service_life = [5, 3, 4]
col = [10, 15, 20]

reactor_life = 4
max_col = 30
```

```python
# Данные вводятся пользователем
# Тип, эффективность, срок службы, количество

types = []
efficiency = []
service_life = []
col = []
reactor_life = 0
max_col = 0

types_col = int(input('Введите число типов стрежней'))

for _ in range(types_col):
    types.append(input('Тип стержня'))
    efficiency.append(float(input('Эффективность')))
    service_life.append(int(input('Срок службы')))
    col.append(int(input('Количество стержней в наличии')))

reactor_life = int(input('Срок службы реактора'))
max_col = int(input('Максимальное количество стержней'))
```

```python
if sum(col) < max_col:
    raise ValueError('Недостаточное количество стержней в наличии')

# Если срок службы меньше, чем у реактора, сразу удаляем
types = [t for t, sl in zip(types, service_life) if sl >= reactor_life]
efficiency = [e for e, sl in zip(
    efficiency, service_life) if sl >= reactor_life]
col = [c for c, sl in zip(col, service_life) if sl >= reactor_life]
service_life = [sl for sl in service_life if sl >= reactor_life]

# Сортируем списки по эффективности
sorted_data = sorted(
    zip(types, efficiency, service_life, col), key=lambda x: x[1], reverse=True)

# Распаковываем отсортированные данные в отдельные списки
types, efficiency, service_life, col = map(list, zip(*sorted_data))

print('Доступные стержни с учетом срока службы реактора:')
print("Тип:", types)
print("Эффективность:", efficiency)
print("Срок службы:", service_life)
print("Количество:", col)

num_final = [0 for _ in range(len(types))]
eff_final = 0.0

# Определяем количество стержней каждого типа
for i in range(len(types)):
    while (num_final[i]+1 <= max_col) and (num_final[i]+1 <= col[i]) and (sum(num_final) < max_col):
        num_final[i] += 1

print('----------------------------------------------------\n')

for i in range(len(types)):
    eff_final += efficiency[i]*num_final[i]


# В алфавитном порядке по типу для вывода
sorted_data = sorted(zip(types, num_final), key=lambda x: x[0])
types, num_final = map(list, zip(*sorted_data))

print('Использованные стержни:')
for i in range(len(types)):
    print(f'Тип {types[i]}: {num_final[i]} стержней')

print(f'Общая эффективность реактора: {eff_final}')
```
